```python
import networkx as nx
import EoN
import matplotlib.pyplot as plt
```

```python
g = nx.read_edgelist("facebook_combined.txt",create_using=nx.Graph(),nodetype=int)
```

```python
print(nx.info(g))
```

```python
spring_pos = nx.spring_layout(g)
```

```python
plt.axis("off")
nx.draw_networkx(g,pos=spring_pos,with_labels=False,node_size=2)
plt.savefig("1.png",dpi=400)
```

```python
sim = EoN.Gillespie_SIR(g,tau = 0.386, gamma = 0.002,return_full_data=True)
```

```python
sim.display(time = 4000,pos=spring_pos,with_labels=False,node_size=5)
plt.savefig("t4000.png",dpi=400)
plt.show()
```

```python
sim.display(time = 8000,pos=spring_pos,with_labels=False,node_size=5)
plt.savefig("13.png",dpi=400)
plt.show()
```

```python
sim.display(time = 1,pos=spring_pos,with_labels=False,node_size=5)
plt.show()
```

```python
sim.display(time = 2,pos=spring_pos,with_labels=False,node_size=5)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(sim.t(), sim.S(), label = "S",color="seagreen")
ax.plot(sim.t(), sim.I(), label = "I",color="red")
ax.plot(sim.t(), sim.R(), label = "R",color="grey")
ax.set_xlabel("Czas")
ax.set_ylabel("Liczebność")
ax.set_xlim([0,10])
ax.set_ylim([3950,4040])
plt.title("Dynamika SIR na sieci rzeczywistej")
plt.legend(loc="best")
#plt.savefig("SIR.png",dpi=400)
plt.show()
```
